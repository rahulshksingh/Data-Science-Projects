```python
from plotnine import *
```

```python
# Create a scatter plot between OperatingCost and TotalSales
```

```python
# Create an empty Canvas
ggplot(data = stores) + theme_gray()
```

```python
# Add axes
ggplot(data = stores) + aes(x = "OperatingCost",y = "TotalSales")
# no quotes in R code
```

```python
# add the actual plot
ggplot(data = stores) + aes(x = "OperatingCost",y = "TotalSales") + geom_point()
```

```python
# adding/modifying labels
(
    ggplot(data = stores) + 
    aes(x = "OperatingCost",y = "TotalSales") + 
    geom_point() +
    xlab("The Expenses - OperatingCost") + 
    ylab("The Income - TotalSales") 
)
 
```

```python
# changing colors
(
    ggplot(data = stores) + 
    aes(x = "OperatingCost",y = "TotalSales") + 
    geom_point(color = "red") +
    xlab("The Expenses - OperatingCost") + 
    ylab("The Income - TotalSales") 
)
 
```

```python
(
    ggplot(data = stores)+
    aes(x = "OperatingCost",y = "TotalSales",color = "Location") + 
    geom_point()
)
```

```python
# Bargraphs

# Location wise TotalSales
temp = stores.groupby(["Location"])
Res = temp[["TotalSales"]].sum().add_prefix("SumOf").reset_index()
Res
```

```python
# Scatter = color = "blue"
# Bar/Hist/Area
#        color = "black" # border color
#        fill = "blue" # fill color
```

```python
ggplot(data = Res) + aes(x = "Location",y = "SumOfTotalSales") + geom_bar(stat = "identity", color = "black", fill = "mediumspringgreen")
```

```python
# Location and StoreType -> Sum of TotalSales
temp = stores.groupby(["Location","StoreType"])
Res = temp[["TotalSales"]].sum().add_prefix("SumOf").reset_index()
Res
```

```python
(
    ggplot(data = Res)+
    aes(x = "Location",y = "SumOfTotalSales", fill = "StoreType")+
    geom_bar(stat = "identity",color = "black")
)
```

```python
(
    ggplot(data = Res)+
    aes(x = "Location",y = "SumOfTotalSales", fill = "StoreType")+
    geom_bar(stat = "identity",color = "black",position = "dodge")+
    geom_text(aes(y = "SumOfTotalSales", label = "SumOfTotalSales"),data = Res)
)
```

```python
# The faceting
# horizontal ~ Vertical

# . ~ Location

(
    ggplot(data = Res)+
    aes(x = "StoreType", y=  "SumOfTotalSales")+
    geom_bar(stat = "identity",fill = "seagreen")+
    facet_grid(".~Location")
)
```
